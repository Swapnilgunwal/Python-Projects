#### Q: Count of transactions where the total amount of transaction was negative.

#### Q: Analyse which product categories are more popular among females vs male customers.

```python
product_g = customer_final.groupby(['Gender','prod_cat'])[['Qty']].sum().reset_index()
product_g.pivot(index = 'prod_cat', columns = 'Gender', values = 'Qty')
```

Products that are popular among males are:

- Books
- Clothing
- Electronics
- Home and Kitchen


Products that are popular among females are:

- Bags
- Footwear

#### Q: Which city code has the maximum customers and what was the percentage of customers from that city.

```python
customer_group = customer_final.groupby('city_code')['customer_Id'].count().sort_values(ascending = False)
customer_group
```

```python
percentage = round((customer_group[4.0]/customer_group.sum()) * 100)
percentage
```

#### Q: Which store type sells the maximum products by value and by quantity.

```python
customer_final.groupby('Store_type')['Qty','Rate'].sum().sort_values(by = 'Qty', ascending = False)
```

e-shop store sells the maximum products by value and by quantity.

#### Q: How many customers have more than 10 unique transactions, after removing all transactions which have any negative counts.

```python
positive_trans = customer_final.loc[customer_final['total_amt'] > 0, :]
unique_trans = positive_trans.groupby(['customer_Id', 'prod_cat','prod_subcat'])['transaction_id'].count().reset_index()
unique_trans_count = unique_trans.groupby('customer_Id')['transaction_id'].count().reset_index()
unique_trans_count[unique_trans_count['transaction_id'] >10 ].count()
```
